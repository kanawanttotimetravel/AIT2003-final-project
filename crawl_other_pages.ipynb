```python
import pandas as pd
from facebook_scraper import get_profile
import time
```

```python
page_lists = pd.read_csv('Data/pages.csv')['0'].to_list()
page_lists
```

```python
profile_list = []
for page in page_lists:
    try:
        profile = get_profile(page, cookies='cookies.txt')
        profile_list.append(profile)
    except Exception as e:
        print(e)
    time.sleep(30)
```

```python
profile_df = pd.DataFrame(profile_list)
profile_df
```

```python
profile_df.to_csv('Data/pages_info.csv')
```
